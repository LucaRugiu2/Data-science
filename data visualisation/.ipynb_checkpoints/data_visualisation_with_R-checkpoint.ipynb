```r
# for this part we need to load an other dataframe
#we will use a csv file with Armstrong wave 4 WAU split by Disco, NWH and Reference cohorts during 2021. The query behind these data is:
setwd("/Users/luca.rugiu/Desktop/Data_science_tools/data visualisation")    #change this into your path to the data
#in che directory sono?
dir()

library(ggplot2)
df2 <- read.csv2("wau_armstrong.csv",header = TRUE,sep=",",na = "NULL")

summary(df2)
```

```r
# box plot (for categorical data)
ggplot(df2, aes(x=TEST_GROUP, y=WAU, fill = TEST_GROUP)) +
  geom_boxplot() +
  theme_classic()+
  scale_fill_manual(values=c("#00796B", "#E69F00", "#8BC34A")) +  #Xing colors
theme(axis.text=element_text(size=10),
        axis.title=element_text(size=10))
```

```r
## barchart with average and standard errors
e <- aggregate(df2$WAU, by = list(TEST = df2$TEST_GROUP), FUN = function(x) c(mean = mean(x), sd = sd(x), n = length(x)))
e <- do.call(data.frame, e)
e$se <- e$x.sd / sqrt(e$x.n)
colnames(e) <- c("TEST", "mean", "sd", "n", "se")
e$names <- c(paste(e$TEST, "TESTS /"))
limits <- aes(ymax = mean + se,ymin = mean - se)
p <- ggplot(data = e, aes(x = factor(TEST), y = mean, fill = factor(TEST)))
```

```r
p + geom_bar(stat = "identity",position = position_dodge(0.9), colour="black") +
  theme(axis.line = element_line(colour = "black"),
        axis.line.x= element_line(colour = "black"),
        axis.line.y= element_line(colour = "black"),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        panel.background = element_blank())+
  geom_errorbar(limits, position = position_dodge(0.9), width = 0.25) +
  labs(x = " ", y = "WAU") +
  ggtitle("Average weekly active users split by Armstrong groups") +
  scale_fill_manual(values=c("#00796B", "#E69F00", "#8BC34A"), name = "Groups")+
  theme(text = element_text(size = 14, family = "sans"))
```

```r
# grouped line plot

ggplot(data=df2, aes(x=WEEK_ID, y=WAU, group=TEST_GROUP)) +
  geom_line(size=1.2)+
  geom_point(size=2) +
geom_line(aes(color=TEST_GROUP))+
  geom_point(aes(color=TEST_GROUP))+
scale_color_manual(values=c("#00796B", "#E69F00", "#8BC34A")) +  #Xing colors
theme_classic()+
theme(axis.text=element_text(size=16),
        axis.title=element_text(size=16))
```

```r
# pie charts
ggplot(df2, aes(x="", y=WAU, fill=TEST_GROUP))+
 geom_bar(width = 1, stat = "identity")+
 coord_polar("y", start=0) +
scale_fill_manual(values=c("#00796B", "#E69F00", "#8BC34A"))+
theme(axis.text=element_text(size=16),
        axis.title=element_text(size=16))+
  theme_minimal()
```

```r
# Density plot with mean lines and marginal rug
# :::::::::::::::::::::::::::::::::::::::::::::::::::
# Change outline and fill colors by groups ("sex")
# Use custom palette
library(ggpubr)
ggdensity(df2, x = "WAU",
   add = "mean", rug = TRUE,
   color = "TEST_GROUP", fill = "TEST_GROUP",
   palette = c("#00796B", "#E69F00", "#8BC34A"))
```

```r
# Box plots with jittered points
# :::::::::::::::::::::::::::::::::::::::::::::::::::
# Change outline colors by groups: dose
# Use custom color palette
# Add jitter points and change the shape by groups
 p <- ggboxplot(df2, x = "TEST_GROUP", y = "WAU",
                color = "TEST_GROUP", palette =c("#00796B", "#E69F00", "#8BC34A"),
                add = "jitter", shape = "TEST_GROUP")

# Add p-values comparing groups
 # Specify the comparisons you want
my_comparisons <- list( c("Armstrong Disco", "Reference"), c("Armstrong NWH", "Reference") )
p + stat_compare_means(comparisons = my_comparisons)+ # Add pairwise comparisons p-value
  stat_compare_means(label.y = 50)   
```

# Deviation graphs
The deviation graph shows the deviation of quantitatives values to a reference value. In the R code below, we’ll plot the JOB_CLICKS z-score from the mtcars dataset.

```r
df <- read.csv2("data2.csv",header = TRUE, sep=",",na = "NULL")
head(df)
```

```r
#Create an ordered barplot, colored according to the level of JOB_CLICKS:
e <- aggregate(df$JOB_CLICKS, by = list(TEST = df$DISCIPLINES), FUN = function(x) c(JOB_CLICKS_Z = mean(x)))
e <- do.call(data.frame, e)
colnames(e) <- c("DISCIPLINES", "JOB_CLICKS")
e$names <- c(paste(e$DISCIPLINES, "DISCIPLINES /"))
e
```

```r
# Calculate the z-score of the variable JOB_CLICKS
e$JOB_CLICKS_z <- (e$JOB_CLICKS - mean(e$JOB_CLICKS))/sd(e$JOB_CLICKS)
e$JOB_CLICKS_grp <- factor(ifelse(e$JOB_CLICKS_z < 0, "low", "high"),
                     levels = c("low", "high"))
e
```

```r
ggbarplot(e, x = "DISCIPLINES", y = "JOB_CLICKS_z",
          fill = "JOB_CLICKS_grp",                        # change fill color by WAU_level
          color = "white",                                # Set bar border colors to white
          palette = c("#00796B", "#E69F00"),              # see ?ggpar
          sort.val = "asc",                               # Sort the value in ascending order
          sort.by.groups = TRUE,                          # Don't sort inside each group
          x.text.angle = 90,                              # Rotate vertically x axis texts
          ylab = "JOB_CLICKS z-score",
          xlab = FALSE,
          legend.title = "JOB_CLICKS Group"
          )
```

```r
#this could be displayed vertically too
ggbarplot(e, x = "DISCIPLINES", y = "JOB_CLICKS_z",
          fill = "JOB_CLICKS_grp",                        # change fill color by WAU_level
          color = "white",                            # Set bar border colors to white
          palette = "jco",            # jco journal color palett. see ?ggpar
          sort.val = "desc",          # Sort the value in descending order
          sort.by.groups = FALSE,     # Don't sort inside each group
          x.text.angle = 90,          # Rotate vertically x axis texts
          ylab = "JOB_CLICKS z-score",
          legend.title = "JOB_CLICKS Group",
          rotate = TRUE,
          ggtheme = theme_minimal()
          )
```

# Lollipop chart
Lollipop chart is an alternative to bar plots, when you have a large set of values to visualize.

Lollipop chart colored by the grouping variable “cyl”:

```r
ggdotchart(e, x = "DISCIPLINES", y = "JOB_CLICKS_z",
           color = "JOB_CLICKS_grp",                        # change fill color by WAU_level                                # Color by groups
           size = 2,
           dot.size = 6,                                  
           palette = c("#00AFBB", "#E7B800", "#FC4E07"), # Custom color palette
           sorting = "ascending",                        # Sort value in descending order
           add = "segments",                             # Add segments from y = 0 to dots
           ggtheme = theme_pubr()                        # ggplot2 theme
           )
```

```r
ggdotchart(e, x = "DISCIPLINES", y = "JOB_CLICKS_z",
           color = "JOB_CLICKS_grp",                                    # Color by groups
           palette = c("#00AFBB", "#E7B800"), # Custom color palette
           sorting = "descending",                       # Sort value in descending order
           add = "segments",                             # Add segments from y = 0 to dots
           rotate = TRUE,                                # Rotate vertically
           group = "JOB_CLICKS",                                # Order by groups
           dot.size = 10,                                 # Large dot size
           label = round(e$JOB_CLICKS),                        # Add mpg values as dot labels
           font.label = list(color = "white", size = 19,
                             vjust = 0.5),               # Adjust label parameters
           ggtheme = theme_pubr()                        # ggplot2 theme
           )
```

```r
ggdotchart(e, x = "DISCIPLINES", y = "JOB_CLICKS_z",
           color = "JOB_CLICKS_grp",                                 # Color by groups
           palette = c("#00AFBB", "#E7B800", "#FC4E07"), # Custom color palette
           sorting = "descending",                       # Sort value in descending order
           rotate = TRUE,                                # Rotate vertically
           dot.size = 6,                                 # Large dot size
           y.text.col = TRUE,                            # Color y text by groups
           ggtheme = theme_pubr()                        # ggplot2 theme
           )+
  theme_cleveland()                                      # Add dashed grids
```
